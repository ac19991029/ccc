# 各列特征解释

数据集有303行，14列，每行表示一个名人。13列表示特征，1列表示标签（是否患有心脏病）。
age：年龄；
sex：性别，1表示男，0表示女；
cp：心绞痛病史 1：典型心绞痛 2：非典型心绞痛 3：无心绞痛 4 ：无症状
trestbps： 静息血压 入院时测量得到，单位为毫米汞柱mm Hg
chol： 胆固醇含量 单位 mg/dl
fbs：空腹时是否血糖高，如果空腹血糖大于120mg/dl 值为1，否则为0
restecg：静息时心电图特征，0：正常；1：ST-T波有异常；2：根据Estes准则，有潜在的左心室肥厚
thalach：最大心率
exang:运动时是否会导致心绞痛 1：会 ；0：不会
oldpeak：运动相比于静息状态。心电图中的ST-T波是否被压平 1：会；0：不会
slope：心电图中ST波峰值坡度 1：上升；2：平坦；3：下降
ca：心脏周边大血管个数（0-3）
thal：是否患有地中海贫血症（3：无；6：fixed defect；7：reversable defect）
target：标签列，是否患心脏病 0：没有；1：有
![jupyter](./1.png)
普通人的ST-T波是平的，如果抬高，说明有异常

# 单列特征与标签的关系

不同年龄段，患心脏病和不患心脏病患者分布

```python
pd.crosstab(df.age,df.target).plot(kind="bar",figsize=(20,6))
plt.title("Heart Disease Frequency for Age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('heartdiseaseandages.png')
plt.show()
```

不同性别，患心脏病和不患心脏病患者分布

```python
pd.crosstab(df.sex,df.target).plot(kind="bar",figsize=(15,6),color=['#1CA53B','#AA1111'])
plt.title("Heart Disease Frequency for Sex")
plt.xlabel('Sex(0=female,1=male)')
plt.xticks(rotation=0)
plt.ylabel('Frequency')
plt.legend(["haven't disease","have disease"])
plt.show()
```

不同slope特征下，患心脏病和不患心脏病患者分布

```python
pd.crosstab(df.slope,df.target).plot(kind="bar",figsize=(15,6),color=['#DAF7A6','#FF5733'])
plt.title("Heart Disease Frequency for Slope")
plt.xlabel('the slope of the peak exercise ST segment')
plt.xticks(rotation=0)
plt.ylabel('Frequency')
plt.show()
```

不同空腹血糖下，患心脏病和不患心脏病患者分布

```python
pd.crosstab(df.fbs,df.target).plot(kind="bar",figsize=(15,6),color=['#FFC300','#581845'])
plt.title("Heart Disease Frequency for FBS")
plt.xlabel('FBS-(fasting blood sugar>120mg/dl)(1=true,0=false)')
plt.xticks(rotation=0)
plt.ylabel('Frequency of disease or not')
plt.legend(["haven't disease","have disease"])
plt.show()
```

不同年龄段，不同最大心率，患心脏病和不患心脏病患者分布

```python
plt.scatter(x=df.age[df.target==1],y=df.thalach[df.target==1],c="red")
plt.scatter(x=df.age[df.target==0],y=df.thalach[df.target==0],c="blue")
plt.legend(["haven't disease","have disease"])
plt.xlabel('Age')
plt.ylabel('maximum heart rate')
plt.show()
#最大心率与患心脏病有关系
```

# 数据预处理

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
```

```python
#查看各列特征数据类型
df.dtypes
```

将简单的列名改为完整的特征名称

```python
df.columns=['age','sex','chest_pain_type','resting_blood_pressure','cholesterol','fasting_blood_sugar','rest_ecg',"max_heart_rate_achieved",'exercise_induced_angina','st_depression','st_slope','num_major_vessels','thalassemia','target']
```

```python
df.head()
```

## 区分 定类 定序 定距 定比 四种特征

```python
#将定类特征由整形编码转化为实际对应的字符串
df['sex'][df['sex']==0]='female'
df['sex'][df['sex']==1]='male'

df['chest_pain_type'][df['chest_pain_type']==0]='typical angina'
df['chest_pain_type'][df['chest_pain_type']==1]='atypical angina'
df['chest_pain_type'][df['chest_pain_type']==2]='non-anginal pain '
df['chest_pain_type'][df['chest_pain_type']==3]='asymptomatic'

df['fasting_blood_sugar'][df['fasting_blood_sugar']==0]='lower than 120mg/ml'
df['fasting_blood_sugar'][df['fasting_blood_sugar']==1]='greater than 120mg/ml'

df['rest_ecg'][df['rest_ecg']==0]='normal'
df['rest_ecg'][df['rest_ecg']==1]='ST-T wave abnormality'
df['rest_ecg'][df['rest_ecg']==2]='left ventricular hypertrophy'


df['exercise_induced_angina'][df['exercise_induced_angina']==0]='no'
df['exercise_induced_angina'][df['exercise_induced_angina']==0]='yes'

df['st_slope'][df['st_slope']==0]='upsloping'
df['st_slope'][df['st_slope']==1]='flat'
df['st_slope'][df['st_slope']==2]='downsloping'

df['thalassemia'][df['thalassemia']==0]='unknow'
df['thalassemia'][df['thalassemia']==1]='normal'
df['thalassemia'][df['thalassemia']==2]='fixed defect'
df['thalassemia'][df['thalassemia']==3]='reversable defect'
```

```python
df.head()
```

```python
#在pandas中 离散的定类和定序特征列应该是 object 对象类型；连续的定居和定比特征列应该是 int64或float64这样的浮点数类型
```

## 将离散的定类和定序特征转为One-Hot独热编码

```python
df=pd.get_dummies(df)
```
